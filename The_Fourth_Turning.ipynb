Authors William Strauss and Neil Howe are widely credited with naming the Millennial generation (see: https://en.wikipedia.org/wiki/Millennials). They are foremost experts and researchers on Generational identities, and have advanced a compelling theory of American history. In one sentence, the theory is that American history has exhibited a cycle, or seasonality, that is intertwined with the lifecycles of the changing generations as they move throughout time. A life, and a Generation lives for about 80-years, so too does the cycle of American history. The authors published a sobering prediction, and the historical precedents for the Crisis we are navigating today.  

### Project Description:

This project seeks to retell their story through data. The data used in this analysis reorganizes America’s political leadership through their generational identities and reorganizes America’s economic trajectory through the lens of Turnings (or "seasons").

The Fourth Turning narrative is incredibly compelling, with anecdotes and facts that help organize the 10,000ft view presented to us.  However, I want to visualize their story through time: What did the prior cycles really look like, and where does that put us today? 

### The Data: 

* American political leadership data includes Presidents and members of Congress from 1789 to the present (2018 – not 2020 yet). The data comes from the @unitedstates project on GitHub, with notable supporters of the database being FiveThirtyEight, GovTrack, ProPublica, and The New York Times. All 67 terms served by presidents (whether the full 4-years or abbreviated) and a total of 41,618 members of Congress are included, although 1,165 (~3%) of Congressional members are missing birthdate information, which prevents any generational analysis. 

The economic data include U.S. GDP-per-Capita, population growth, and Federal Debt-to-GDP. 
* The measure of real GDP per capita – adjusting for inflation – comes from the venerable Maddison Project Database, part of University of Groningen, Netherlands. Real GDP-per-Capita is arguably the most prevalent development statistic used to measure a nation – capturing the scale of output and income. The data is available in a 1-year interval beginning in 1800, so that is what is used. The Maddison Project Database is referenced by publications out of Oxford, the University of Chicago, and Stanford (see the links below for reference). 

* Data covering the U.S. Federal Government’s Debt-to-GDP from 1790 through today comes from the U.S. Congressional Budget Office. While Debt-to-GDP does not capture the entire store of social liabilities - there are health care burdens, pension burdens, and corporate and consumer debt - it does tell a large part of the story, and one that is directly impacted by the decisions of America's leadership.

### American history follows a cyclicality, a seasonality...
The theory is that history, specifically American history, follows a cyclical pattern, analogous in many ways to nature's seasons: Winter, Spring, Summer, and Fall. The Winter is a Crisis period – which has coincided with the major wars of the American Revolution, the Civil War, and World War II. Spring is the post-war boom (ex. 1950’s), Autumn is the pre-war unraveling, and Summer is a cultural awakening (ex. the counter-culture revolution of the 1960's and American history's major religious "awakenings": the First Great Awakening 1730's, Second Great Awakening in 1790's, and the Third Great Awakening in the 1890's).  

### ...As each generation has a unique persona, which is shaped by history in their youth...
These four seasons of history span the length of a typical generation (about 20 years) and combine to about 80-years, the span of a human life. The phenomenon of generations and their unique identities, is fundamentally intertwined with this historical seasonality and cyclicality. Generational identities are shaped by the events of history in their youth years, and later in life they, in-turn shape the events of history. 

### ...And shapes history in adulthood.
A pattern emerges. As each generation ages and transitions to the next stage of life (Child to Young Adult, to Midlife, to Elderhood), history too transitions to the next phase of the cycle, the next season of history. And while history tends to look very different from one season to the next, just as a generation tends to look very different from one to the next, history does tends to rhyme when viewed from a longer timeline.

```python
# Convert YAML to DataFrame
def yaml_to_df(yaml_data):
    data_list = []
    for index, entry in enumerate(yaml_data):
        for term, entry_terms in enumerate(entry['terms']):
            try:
                _birthday = entry['bio']['birthday']
            except KeyError:
                _birthday = 'NaT'
            try:
                _party = entry_terms['party']
            except KeyError:
                _party = 'NaN'
            
            data_list.append({
                'Name': ('_'.join([entry['name']['first'], entry['name']['last']])),
                'Birthday': _birthday,
                'Job': entry_terms['type'],
                'Start': entry_terms['start'],
                'End': entry_terms['end'],
                'Party': _party,
            })
    data_df = pd.DataFrame(data_list)
    return data_df
```

```python
import re

period_regex = re.compile('(.*)\s\((\d{4}).(\d{4}).?\)')  # '(.*)\s\((\d{4}).(\d{4}).?\)'
# Resource: https://regex101.com/

turnings_list = [];  # saeculum_list = []
for turning, _periods in _period_df.iteritems():
    for saeculum, _string in _periods.iteritems():
        _regex_match = re.search(period_regex, _string)
        if _regex_match is not None:
            period, start, end = _regex_match.groups()
            turnings_list.append({'Turning': turning, 'Period': period.rstrip(), 'Saeculum': saeculum, 
                                  'Start': start, 'End': end})
        pass
    pass

turnings_df = pd.DataFrame(turnings_list)
```

```python
# Scoring: Executive

avg_score_df = pd.DataFrame()
first_score_2_df = pd.DataFrame()
for i in range(4):  # Loop through Turnings 
    pivot_table = exec_pivot_turnings_dict[turnings_keys[i]]

    # 1) Values:
    _turning_score_details_df = pd.DataFrame(pivot_table.mean(), columns=[turnings_keys[i]])
    _mean = pd.Series(_turning_score_details_df.mean(), name='Mean')
    _var = pd.Series(_turning_score_details_df.var(), name='Var')
    _count = pd.Series(_turning_score_details_df.count(), name='Count')
    avg_score_df = avg_score_df.append(_mean.to_frame().join(_var).join(_count))

    # 2) First score >= 2:
    # First, Calc stats accross each period
    first_score_2_dict = {}
    for column in list(pivot_table.columns):
        first_score_2_dict[column] = pivot_table.index[pivot_table[column] >= 2].min()
    # Second, Calc stats accross each turning
    _turning_score_details_df = pd.DataFrame.from_dict(first_score_2_dict, orient='index', 
                                                       columns= [turnings_keys[i]])
    _mean = pd.Series(_turning_score_details_df.mean(), name='Mean')
    _var = pd.Series(_turning_score_details_df.var(), name='Var')
    _count = pd.Series(_turning_score_details_df.count(), name='Count')
    first_score_2_df = first_score_2_df.append(_mean.to_frame().join(_var).join(_count))

# Join score stats together into a dictionary:
exec_score_dict = {'Average_Score': avg_score_df, 
                   'First_Year_2': first_score_2_df}
```

```python
# Means Test to see if the implied distributions are statistically different:

# From Scipy:
# This is a two-sided test for the null hypothesis that two independent samples have identical average 
# (expected) values. Perform Welch’s t-test, which does not assume equal population variance
# From Wiki:
# The p-value is the probability of obtaining test results at least as extreme as the results actually observed,
# under the assumption that the null hypothesis is correct. 
# A very small p-value means that such an extreme observed outcome would be very unlikely 
# under the null hypothesis.

# T-test: 
from scipy.stats import ttest_ind_from_stats

def score_ttest(score_df):  # _score_name
    #pvalue_df = pd.DataFrame()
    #t_statistic_df = pd.DataFrame()
    #for _score in list(score_dict.keys()):
    #    score_df = score_dict[_score]
    #_score = 'Average_Score'
    #score_df = exec_score_dict[_score]
    mean_test = [[0,1],  # First-turning vs. Second-turning
                 [1,2],  # Second-Turning vs. Third-turning
                 [2,3],  # Third-turning vs. Fourth-turning
                 [0,2],  # First-turning vs. Third-turning
                 [0,3],  # First-turning vs. Fourth-turning
                 [1,3]]  # Second-Turning vs. Fourth-turning
    pvalue_dict = {}; t_statistic_dict = {}
    for i, j in mean_test:
        t_statistic, pvalue = ttest_ind_from_stats(mean1=score_df['Mean'][i], std1=np.sqrt(score_df['Var'][i]),
                                                   nobs1=score_df['Count'][i],
                                                   mean2=score_df['Mean'][j], std2=np.sqrt(score_df['Var'][j]),
                                                   nobs2=score_df['Count'][j],
                                                  equal_var=False)
        pvalue_dict.update({"{}_vs_{}".format(score_df.index[i], score_df.index[j]): pvalue})
        t_statistic_dict.update({"{}_vs_{}".format(score_df.index[i], score_df.index[j]): t_statistic})

    pvalue_df = pd.DataFrame.from_dict(pvalue_dict, orient='index', columns=['P_Value'])
    t_statistic_df = pd.DataFrame.from_dict(t_statistic_dict, orient='index', columns=['t_Stat'])
    score_summary_df = pvalue_df.join(t_statistic_df)
    return score_summary_df
```

```python
# Scoring: Congress

def score_statistics(pivot_turnings_dict, turnings_keys, key_level):
    
    avg_score_df = pd.DataFrame()
    first_score_percent_df = pd.DataFrame()
    for i in range(4):  # Loop through Turnings 
        pivot_table = pivot_turnings_dict[turnings_keys[i]]

        # 1) Values:
        _turning_score_details_df = pd.DataFrame(pivot_table.mean(), columns=[turnings_keys[i]])
        _mean = pd.Series(_turning_score_details_df.mean(), name='Mean')
        _var = pd.Series(_turning_score_details_df.var(), name='Var')
        _count = pd.Series(_turning_score_details_df.count(), name='Count')
        avg_score_df = avg_score_df.append(_mean.to_frame().join(_var).join(_count))

        # 2) First score >= 2:
        # First, Calc stats accross each period
        first_score_dict = {}
        for column in list(pivot_table.columns):
            first_score_dict[column] = pivot_table.index[pivot_table[column] >= key_level].min()
        # Second, Calc stats accross each turning
        _turning_score_details_df = pd.DataFrame.from_dict(first_score_dict, orient='index', 
                                                           columns= [turnings_keys[i]])
        _mean = pd.Series(_turning_score_details_df.mean(), name='Mean')
        _var = pd.Series(_turning_score_details_df.var(), name='Var')
        _count = pd.Series(_turning_score_details_df.count(), name='Count')
        first_score_percent_df = first_score_percent_df.append(_mean.to_frame().join(_var).join(_count))

    # Join score stats together into a dictionary:
    summary_score_dict = {'Average_Score': avg_score_df, 'First_Year_Key_Level': first_score_percent_df}
    return summary_score_dict
```
